```kotlin
//简单数据集示例
val simpleDataset = mapOf(
    "time, ms" to listOf(12, 87, 130, 149, 200, 221, 250),
    "relativeHumidity" to listOf(0.45, 0.3, 0.21, 0.15, 0.22, 0.36, 0.8),
    "flowOn" to listOf(true, true, false, false, true, false, false)
)
```

```kotlin
// 1. 使用具有指定列类型和名称的 `column()` 函数:
val timeMs = column<Int>("time, ms")
// 2. 利用 String API，与上面的方法类似，但使用 String 调用:
val humidity = "relativeHumidity"<Double>()
// 3. 委托一个未命名的列 - 其名称将从变量名称派生而来:
val flowOn by column<Boolean>()
```

```kotlin
// 创建具有自定义轴和图例的绘图
plot(simpleDataset) {
    points {
        // 设置x轴标题
        x(timeMs) {
            axis.name = "Time from start of counting,\n milliseconds"
        }
        // 设置y轴标题和样式
        y(humidity) {
            scale = continuous(0.0..1.0) // 湿度指标为[0.0,1.0]
            axis {
                name = "Relative humidity" // 坐标轴名称
                breaksLabeled(0.0 to "0%", 0.3 to "30%", 0.6 to "60%", 0.9 to "90%") // 轴线和标签
            }
        }
        size = 12.0 // 点尺寸
        // 配置湿度图例
        color(humidity) {
            scale = continuous()
            legend {
                name = "rel. humidity" // 图题
                type = LegendType.ColorBar(40.0, 190.0, 15) // 图例类型和尺寸
                 // breaks(format = "e")   // 格式化数字为科学计数法
            }
        }
    }
}
```

```kotlin
plot(simpleDataset) {
    points {
        x(timeMs)// 设置x轴

        y(humidity)// 设置y轴

        size = 4.5// 点尺寸

        color(flowOn)// 根据值选择颜色
    }
}
```
